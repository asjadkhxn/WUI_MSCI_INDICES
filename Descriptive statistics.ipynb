```python
import pandas as pd
from scipy.stats import skew, kurtosis
from statsmodels.tsa.stattools import adfuller
from scipy import stats
from statsmodels.stats.diagnostic import acorr_ljungbox 

# Calculate descriptive statistics of the data, including skewness and kurtosis
data2=pd.DataFrame({'Count': data.count(),
                    'Mean': data.mean(),'Max': data.max(),
                    'Min': data.min(),                   
                     'Std': data.std(),
                    'Skewness': skew(data),
                    'Kurtosis': kurtosis(data)})

data2.to_excel('Descriptive statistics.xlsx',sheet_name='energy_basic_characteristics')

import pandas as pd  
from scipy.stats import jarque_bera  
from statsmodels.tsa.stattools import adfuller  
from statsmodels.stats.diagnostic import acorr_ljungbox  
  
results_list = []  
  

for i in range(data.shape[1]):  
    
    column_data = data.iloc[:, i]  
      
    # Jarque-Bera test 
    jb_stat, jb_p = jarque_bera(column_data)  
      
    # ADF test
    result = adfuller(column_data)  
      
    # Ljung-Box Q statistic test
    result_q10 = acorr_ljungbox(column_data, lags=[10], boxpierce=False)  
      
    
    results_list.append({  
        'ADF': result[0],  
  
        'J-B_stat': jb_stat,  

        'Q(10)': result_q10.iloc[0,0], 
                'ADF_p': result[1],
                'J-B_p': jb_p,  
        'Q(10)_p': result_q10.iloc[0,1]  
    })  

results_df = pd.DataFrame(results_list)  

with pd.ExcelWriter('Descriptive statistics.xlsx', engine='openpyxl', mode='a') as writer:
    results_df.to_excel(writer, sheet_name='Energy Smoothing Test')
```
